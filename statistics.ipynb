```python
import os
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams.update({'font.size': 12})
```

```python
TRAIN_IMAGES_FOLDER = './dataset/images/train'
TRAIN_LABELS_FOLDER = './dataset/labels/train'

TEST_IMAGES_FOLDER  = './dataset/images/test'
TEST_LABELS_FOLDER  = './dataset/labels/test'
```

# 1. Thống kê số lượng ảnh trong tập dữ liệu

```python
train_images_file   = os.listdir(TRAIN_IMAGES_FOLDER)
test_images_file    = os.listdir(TEST_IMAGES_FOLDER)

fig, ax = plt.subplots()
bar_container = ax.bar(['train', 'test'], [len(train_images_file), len(test_images_file)])
ax.bar_label(bar_container, padding=-50, color='w', weight='bold')
ax.set_ylabel('Số lượng ảnh')
plt.show()
```
